```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.metrics import mean_squared_error

try:
    train = pd.read_csv('path/to/train.csv')
    test = pd.read_csv('path/to/test.csv')
except FileNotFoundError:
    print("Update file paths for train.csv and test.csv")
    train = pd.DataFrame({
        'ID': range(100),
        'date': pd.date_range('2022-01-01', periods=100),
        'Item Id': np.random.randint(1, 5, 100),
        'Item Name': ['A']*100,
        'anarix_id': ['X']*100,
        'units': np.random.randint(10, 50, 100),
        'unit_price': np.random.uniform(5, 15, 100),
        'ad_spend': np.random.uniform(0, 10, 100)
    })
    test = pd.DataFrame({
        'ID': range(100, 120),
        'date': pd.date_range('2022-04-11', periods=20),
        'Item Id': np.random.randint(1, 5, 20),
        'Item Name': ['A']*20,
        'anarix_id': ['X']*20,
        'unit_price': np.random.uniform(5, 15, 20),
        'ad_spend': np.random.uniform(0, 10, 20)
    })

test_ids = test['ID']

train['date'] = pd.to_datetime(train['date'], errors='coerce')
test['date'] = pd.to_datetime(test['date'], errors='coerce')

train = train[train['unit_price'] > 0].copy()
train['units'] = train['units'].clip(lower=0)

train['Item Id'].fillna(train['Item Id'].mode()[0], inplace=True)
test['Item Id'].fillna(test['Item Id'].mode()[0], inplace=True)
train['ad_spend'].fillna(0, inplace=True)
test['ad_spend'].fillna(0, inplace=True)

test['units'] = np.nan
data = pd.concat([train, test], ignore_index=True).sort_values(['Item Id', 'date'])

data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
data['weekday'] = data['date'].dt.weekday

for lag in range(1, 8):
    data[f'units_lag_{lag}'] = data.groupby('Item Id')['units'].shift(lag)

data['units_roll_mean_7'] = data.groupby('Item Id')['units'].transform(
    lambda x: x.shift(1).rolling(7).mean()
)

data.dropna(subset=[c for c in data if 'lag' in c or 'roll' in c], inplace=True)

train_data = data[~data['ID'].isin(test_ids)].copy()
test_data = data[data['ID'].isin(test_ids)].copy()

split_date = train_data['date'].max() - pd.DateOffset(days=30)
train_idx = train_data['date'] <= split_date
val_idx = train_data['date'] > split_date

features = [c for c in train_data.columns if c not in ['date', 'units', 'Item Name', 'anarix_id', 'ID']]
target = 'units'

X_train, y_train = train_data.loc[train_idx, features], train_data.loc[train_idx, target]
X_val, y_val = train_data.loc[val_idx, features], train_data.loc[val_idx, target]

model = lgb.LGBMRegressor(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_val)
mse = mean_squared_error(y_val, y_pred)
print(f"RMSE: {np.sqrt(mse):.4f}")

final_model = lgb.LGBMRegressor(random_state=42)
final_model.fit(train_data[features], train_data[target])

X_test = test_data[features]
preds = final_model.predict(X_test)

preds = np.clip(preds, 0, None).round().astype(int)

submission = pd.DataFrame({'ID': test_data['ID'], 'TARGET': preds})
submission.to_csv('submission.csv', index=False)

print("submission.csv created")
```
